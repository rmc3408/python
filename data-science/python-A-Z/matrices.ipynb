Custom Plotting by player function

```python
def showPlotByNames(data, playerList = ['LeBronJames']):
  x = Seasons
  colors = ['tab:orange', 'tab:brown', 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

  for player in playerList:
    plotLine = plt.plot(x, data[Pdict[player]])
    plt.setp(plotLine, linewidth=2, color=colors[Pdict[player]], linestyle='-', marker='o', markersize=6, label=player)

  plt.legend()
  plt.xlabel('Seasons')
  plt.show()
```

```python
showPlotByNames(realScores, ['JoeJohnson', 'KevinDurant', 'ChrisBosh'])
showPlotByNames(Salary, ['JoeJohnson', 'KevinDurant', 'ChrisBosh'])
```

```python
showPlotByNames(Salary, ['JoeJohnson', 'KevinDurant'])
showPlotByNames(Salary / Games, ['JoeJohnson', 'KevinDurant'])
showPlotByNames(Salary / realScores, ['JoeJohnson', 'KevinDurant'])
```
